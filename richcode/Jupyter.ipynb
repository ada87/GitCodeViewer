# Task Management Analysis

This Jupyter Notebook demonstrates data analysis for task management.

```python
import json

# Sample task data
tasks = [
    {"id": 1, "title": "Implement auth", "priority": "high", "status": "in_progress"},
    {"id": 2, "title": "Write tests", "priority": "medium", "status": "pending"},
    {"id": 3, "title": "Fix bug", "priority": "critical", "status": "completed"},
]

print(f"Total tasks: {len(tasks)}")
```

## Task Statistics

```python
# Count by status
from collections import Counter

status_counts = Counter(task["status"] for task in tasks)
priority_counts = Counter(task["priority"] for task in tasks)

print("By Status:")
for status, count in status_counts.items():
    print(f"  {status}: {count}")

print("\nBy Priority:")
for priority, count in priority_counts.items():
    print(f"  {priority}: {count}")
```
